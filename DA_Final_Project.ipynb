```python
import pandas as pd
import numpy as np
def check_missing_values_flexible(df, column):
    """
    Check missing values in a column, including:
    - np.nan / None
    - empty strings ""
    - string representations like "null", "NULL", "None"
    Works for numeric and string columns.
    """
    null_strings = ['null', 'none', '']  # all lowercase
    col_data = df[column]

    # Count actual NaN / None
    nan_count = col_data.isna().sum()

    # For object/string columns, count null-like strings (case-insensitive)
    str_null_count = 0
    if col_data.dtype == 'O':  # object type
        str_null_count = col_data.astype(str).str.strip().str.lower().isin(null_strings).sum()

    total_missing = nan_count + str_null_count
    percentage_missing = total_missing / len(df) * 100

    print("="*60)
    print(f"Missing Value Summary for Column: {column}")
    print("="*60)
    print(f"Total Rows: {len(df)}")
    print(f"Actual NaN/None values: {nan_count}")
    print(f"Null-like strings / empty strings: {str_null_count}")
    print(f"Total Missing Values: {total_missing}")
    print(f"Percentage Missing: {percentage_missing:.2f}%")
    print("="*60)

    return {
        'nan_count': nan_count,
        'null_string_count': str_null_count,
        'total_missing': total_missing,
        'percentage_missing': percentage_missing
    }
```

```python
# =============================================================
# Cleaning and Inspecting 'bmi' Columns
# =============================================================

import pandas as pd

def check_outliers_iqr(df, column):
    """
    Detect outliers in a numeric column using the IQR (Interquartile Range) method.

    Parameters:
    df (pd.DataFrame): The input DataFrame.
    column (str): The column name to analyze.

    Returns:
    dict: Summary of statistics and outlier thresholds.
    """

    print("="*60)
    print(f"Outlier Detection for Column: {column}")
    print("="*60)

    # Drop NA to avoid errors in quantile calculation
    col_data = df[column].dropna()

    # Calculate key statistics
    min_val = col_data.min()
    q1 = col_data.quantile(0.25)
    q2 = col_data.quantile(0.50)
    q3 = col_data.quantile(0.75)
    max_val = col_data.max()

    # Calculate IQR
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    upper_limit = q3 + 1.5 * iqr

    # Detect outliers
    outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]

    # Print summary
    print(f'''
Minimum Value      : {min_val}
1st Quartile (Q1)  : {q1}
Median (Q2)        : {q2}
3rd Quartile (Q3)  : {q3}
Maximum Value      : {max_val}
---------------------------------------------
IQR Value          : {iqr}
Lower Limit        : {lower_limit}
Upper Limit        : {upper_limit}
Outliers Detected  : {len(outliers)}
''')

    return {
        "min": min_val,
        "q1": q1,
        "median": q2,
        "q3": q3,
        "max": max_val,
        "IQR": iqr,
        "lower_limit": lower_limit,
        "upper_limit": upper_limit,
        "outliers_count": len(outliers),
        # "outlier_rows": outliers
    }


print(check_missing_values_flexible(df,"bmi"),"null values check")

check_outliers_iqr(df,"bmi")
```
